```python
# Memilih fitur untuk pengelompokan dari dataframe
X1 = df[['Tingkat Kerusakan' , 'Tingkat Dampak']].iloc[: , :].values
#inertia untuk metrik  mengukur sejauh mana titik-titik dalam suatu klaster berjarak dari pusat klasternya
inertia = []
#Melakukan iterasi melalui berbagai jumlah klaster (n) dari 1 hingga 10
for n in range(1 , 11):
   # Menginisialisasi algoritma KMeans dengan parameter-parameter tertentu
  algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300,
                      random_state= 111) )
  # Melatih algoritma KMeans pada data
  algorithm.fit(X1)
   # Menambahkan nilai inersia ke dalam list
  inertia.append(algorithm.inertia_)
```

```python
# Membuat sebuah figure dengan ukuran 15x6 inci
plt.figure(1 , figsize = (15 ,6))
# Plotting inersia terhadap jumlah klaster
plt.plot(np.arange(1 , 11) , inertia , 'o') # Menandai titik-titik
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5) # Menghubungkan titik-titik dengan garis
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

```python
# bangun K-Means
# Inisialisasi algoritma KMeans
algorithm = (KMeans(n_clusters = 2 ,init='k-means++', n_init = 10 ,max_iter=300,
                    tol=0.0001, random_state= 111 , algorithm='elkan') )
# Melatih algoritma KMeans pada data X1
algorithm.fit(X1)
# Mendapatkan label klaster untuk setiap titik data
labels2 = algorithm.labels_
# Mendapatkan koordinat pusat klaster
centroids2 = algorithm.cluster_centers_
```

```python
# Mendapatkan label klaster untuk setiap titik data
labels2 = algorithm.labels_
# Mendapatkan koordinat pusat klaster
centroids2 = algorithm.cluster_centers_
# Menentukan langkah (step) untuk membuat meshgrid
step = 0.02
# Menentukan batas x dan y dari data
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
# Membuat meshgrid
xx, yy = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
# Memprediksi label klaster untuk setiap titik dalam meshgrid
Z1 = algorithm.predict(np.c_[xx.ravel(), yy.ravel()]) # array diratakan 1D
```

```python
plt.figure(1 , figsize = (15 , 7) )
plt.clf()
# Mengubah bentuk Z1 agar sesuai dengan ukuran meshgrid
Z1 = Z1.reshape(xx.shape)
# Menampilkan peta warna dari hasil klastering (Z1)
plt.imshow(Z1 , interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap = plt.cm.Pastel2, aspect = 'auto', origin='lower')
# Menampilkan scatter plot dari data dengan warna berdasarkan label klaster (labels2)
plt.scatter( x ='Tingkat Kerusakan'  ,y = 'Tingkat Dampak' , data = df , c = labels2 ,s = 200 )
# Menampilkan pusat klaster dengan warna merah
plt.scatter(x = centroids2[: , 0] , y = centroids2[: , 1] , s = 300 , c = 'red' ,
            alpha = 0.5)
plt.ylabel('Tingkat Dampak') , plt.xlabel('Tingkat Kerusakan')
plt.show()
```

```python
from flask import Flask, request, jsonify
app = Flask(__name__)

@app.route('/', methods=['GET'])
def index():
   return "<h1>Hello this is Data Scsience</h1>"

# Endpoint untuk melakukan prediksi berdasarkan data yang diterima
@app.route('/predict', methods=['GET', 'POST'])
def upload():
  # Inisialisasi variabel data dengan nilai default
  data={"success":False}
   # Memeriksa jenis metode permintaan (GET atau POST)
  if request.method == 'POST':
    # Mengambil data 'Tingkat Kerusakan' dan 'Tingkat Dampak' dari formulir
    kerusakan = request.form['Tingkat Kerusakan']
    dampak = request.form['Tingkat Dampak']
    # Membuat array data dari nilai-nilai yang diterima
    data = [[kerusakan, dampak]]

    try:
       # Melakukan prediksi menggunakan model (algorithm)
      prediksi = algorithm.predict(data)
      prediksi_list = prediksi.tolist()
      print(prediksi_list)
      # Mengembalikan hasil prediksi dalam format JSON
      return jsonify(prediksi_list)
    except Exception as e:
      # Mengembalikan pesan kesalahan jika terjadi error selama prediksi
        return jsonify({"error": str(e)})
  else:
    # Mengembalikan pesan jika metode permintaan bukan POST
     return "<h1>Method Salah</h1>"
```
